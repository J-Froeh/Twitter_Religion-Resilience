## Prep right Tweet set

```python
# In this section keywords are loaded used to identify tweets related to the pandemic
# Having the keywords loaded the tweets are filtered and marked whether they are related to the pandemic or not
# Lastly, the dataset is reduced to communication that is Covid-related
```

```python
def prepare_keywords(file):
    keys = pd.read_csv(file)
    s = ''
    return [(s+word) for word in keys['keywords']]

def covid_relation(df, keylist):
    return df.assign(covid_related=df['text_processed'].apply(lambda t: 1 if any((' ' + keyword + ' ') in ' '.join(t) for keyword in keylist) else 0))

def no_covid_relation(df, keylist):
    return df.assign(not_covid_related=df['covid_related'].apply(lambda t: 0 if t==1 else 1))
```

```python
keylist = prepare_keywords('keywords.txt')
keylist
```

```python
tdf = (tdf
       .drop(columns=['author_id', 'created_truncated', 'referenced_type',
                      'entities.mentions', 'entities.annotations', 'entities.hashtags',
                      'is_reply', 'length', 'public_metrics.followers_count'])
       .pipe(covid_relation, keylist)
       .pipe(no_covid_relation, keylist)
      )
tdf
```

```python
df = (tdf[tdf['covid_related']==1]
      .drop(columns=['created_at', 'tweet_count', 'name', 'country',
                     'covid_related', 'not_covid_related'])
      .rename(columns={'text_processed': 'tokens'})
      )
df
```

## Cleanup preprocessing

```python
# The Covid-communication is prepared for topic-modeling by a) removing URLs and
# b) removing the keywords used to identify Covid-relation
# (by being the selection criteria for the dataset these words dont convey any further insights)
```

```python
def remove_url(token_list):
    tokens = (token for token in token_list if token != '<-url->')
    return list(tokens)
```

```python
# Removing top 5 - words used for identifying Covid-relation
# Additionally, all spellings of "covid" and "corona" within the top 15 most used terms are removed
def remove_covid_keys(token_list):
    key_list = ['#covid19', 'covid', 'pandemic', 'coronavirus', '#coronavirus',
               '#covid_19', '#pandemic', '#covid', '#covid-19']
    tokens = (token for token in token_list if token not in key_list)
    return list(tokens)
```

```python
df['tokens'] = (df['tokens']
                .apply(lambda t: remove_url(t))
                .apply(lambda t: remove_covid_keys(t))
               )
df
```

## Bigrams and Trigrams

```python
# Bigram and Trigram models are generated and then common n-grams in the dataset identified before topic modeling
```

```python
# Build the bigram and trigram models
bigram = gensim.models.Phrases(df['tokens'], min_count=5, threshold=100) # higher threshold fewer phrases.
trigram = gensim.models.Phrases(bigram[df['tokens']], threshold=100)

# Faster way to get a sentence clubbed as a trigram/bigram
bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)
```

```python
def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]

def make_trigrams(texts):
    return [trigram_mod[bigram_mod[doc]] for doc in texts]
```

```python
df['tokens_N_grams'] = make_trigrams(df['tokens'])
df
```

## Lemmatization

```python
# Using the spacy-library, the tweets are lemmatized and prepared for topic modeling
```

```python
# Refrence 4 : https://stackoverflow.com/questions/45306988/column-of-lists-convert-list-to-string-as-a-new-column
nlp = spacy.load('en_core_web_lg')

# Timing Start
program_start_time = time.time()

# Make tokens a string again
df['tokens_back_to_text'] = [' '.join(map(str, l)) for l in df['tokens_N_grams']]

def get_lemmas(text):
    '''Used to lemmatize the processed tweets'''
    lemmas = []
    
    doc = nlp(text)
    
    # Something goes here :P
    for token in doc: 
        if ((token.is_stop == False) and (token.is_punct == False)) and (token.pos_ != 'PRON'):
            lemmas.append(token.lemma_)
    
    return lemmas

df['lemmas'] = df['tokens_back_to_text'].apply(get_lemmas)

# Make lemmas a string again
df['lemmas_back_to_text'] = [' '.join(map(str, l)) for l in df['lemmas']]
# df[['original_tweet', 'lemmas_back_to_text']]

# Timing End
program_end_time = time.time()
```

```python
#Printing Lemmetization Time
print(program_end_time - program_start_time, "seconds to finish")
```

```python
# Timing Start
program_start_time = time.time()

# Tokenizer
tokenizer = Tokenizer(nlp.vocab)

# Tokenizer function
def tokenize(text):
    """
    Parses a string into a list of semantic units (words)
    Args:
        text (str): The string that the function will tokenize.
    Returns:
        list: tokens parsed out
    """
    # Removing url's
    pattern = r"http\S+"
    
    tokens = re.sub(pattern, "", text) # https://www.youtube.com/watch?v=O2onA4r5UaY
    tokens = re.sub('[^a-zA-Z 0-9]', '', text)
    tokens = re.sub('[%s]' % re.escape(string.punctuation), '', text) # Remove punctuation
    tokens = re.sub('\w*\d\w*', '', text) # Remove words containing numbers
    tokens = re.sub('@*!*\$*', '', text) # Remove @ ! $
    tokens = tokens.strip(',') # TESTING THIS LINE
    tokens = tokens.strip('?') # TESTING THIS LINE
    tokens = tokens.strip('!') # TESTING THIS LINE
    tokens = tokens.strip("'") # TESTING THIS LINE
    tokens = tokens.strip(".") # TESTING THIS LINE

    tokens = tokens.lower().split() # Make text lowercase and split it
    
    return tokens

# Apply tokenizer
df['lemma_tokens'] = df['lemmas_back_to_text'].apply(tokenize)

# Timing End
program_end_time = time.time()

# View those tokens (the 4th column)
df
```

```python
#Printing Tokenization Time
print(program_end_time - program_start_time, "seconds to finish")
```

## Topic Modeling

```python
# In this first section, the dictionary and corpus for the topic model are created
# Afterwards, the optimized model generated previously is loaded
```

##### id2word

```python
# Create a id2word dictionary
id2word = Dictionary(df['lemma_tokens'])
print(len(id2word))
```

```python
# Filtering Extremes
id2word.filter_extremes(no_below=2, no_above=.99)
print(len(id2word))
```

##### Corpus Object & Generating Base Model Topics

```python
# Creating a corpus object 
corpus = [id2word.doc2bow(d) for d in df['lemma_tokens']]
```

##### Setting up logger

### Load finale model

```python
# Load a potentially pretrained model from disk.
model_final = datapath("model_final")
model_5_1 = LdaMulticore.load(model_final)
model_5_1
```

```python
# Compute Perplexity
# a measure of how good the model is. lower the better
model_5_1_perplexity = model_5_1.log_perplexity(corpus)
print('\nPerplexity: ', model_5_1_perplexity) 

# Compute Coherence Score
coherence_model_5_1 = CoherenceModel(model=model_5_1, texts=df['lemma_tokens'], 
                                   dictionary=id2word, coherence='c_v')
coherence_lda_model_5_1 = coherence_model_5_1.get_coherence()
print('\nCoherence Score: ', coherence_lda_model_5_1)
```
